```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
screen_time_hours = pd.Series(
    [
        8,
        7,
        8,
        9,
        9,
        8,
        7,
        5,
        8,
        6,
        6,
        6,
        8,
        5,
        7,
        6,
        3,
    ]
)
print(screen_time_hours.value_counts())
print(len(screen_time_hours))
```

```python
sns.histplot(x=screen_time_hours, bins=len(np.unique(screen_time_hours)))
plt.title("Laj's Phone Screen Time")
plt.xlabel("Hours")
plt.show()
```

```python
sns.boxplot(x=screen_time_hours)
plt.xlabel("Hours")
plt.title("Laj's Phone Screen Time")
plt.show()
```
